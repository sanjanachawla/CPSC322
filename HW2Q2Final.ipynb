```python
#homework 2 q2
# DSP with Pruning


def check_constraints(A,B,C,D,E,F,G,H, Dom):
    if((not A>=G) and G in Dom and A in Dom):
        return False
    elif(not(abs(G-C)==1) and G in Dom and C in Dom):
        return False
    elif( not (D!=C) and D in Dom and C in Dom):
        return False
    elif( not (G!=F) and G in Dom and F in Dom):
        return False
    elif( not (abs(E-F)%2 == 1) and E in Dom and F in Dom):
        return False
    elif( not(A<H) and H in Dom and A in Dom):
        return False
    elif (not (abs(H-C)%2 ==0) and H in Dom and C in Dom):
        return False
    elif (not(E!=C) and E in Dom and C in Dom):
        return False
    elif( not(H!=F) and H in Dom and F in Dom):
        return False
    elif (not(abs(F-B)==1) and F in Dom and B in Dom):
        return False
    elif (not(H!=D) and H in Dom and D in Dom):
        return False
    elif (not(E< D-1) and E in Dom and D in Dom):
        return False
    elif (not(C!=F) and C in Dom and F in Dom):
        return False
    elif (not(G<H) and G in Dom and H in Dom):
        return False
    elif(not(D>=G) and D in Dom and G in Dom):
        return False
    elif (not(E!=H-2) and E in Dom and H in Dom):
        return False
    elif(not(D!=F) and D in Dom and F in Dom):
        return False
    else:
        return True
```

```python
#for any List
#init all variables to 0

ordering = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'] #order of vars
Dom = [1,2,3,4] #domain
path = [] #current working path in order of vars
sols = [] #to keep track of all complete solutions
solutions = []
#init all vars to 0 (note: this is outside Dom)
A=0;
B=0;
C=0;
D=0;
E=0;
F=0;
G=0;
H=0;

i = 0 
sols_found = 0 
fails_found = 0

while i in range(0, len(Dom)): #iterate through domains
    j = 0
    while(j < len(ordering)): #iterate through variables
                   
        V = ordering[j] #variable to keep track of current var
        d = Dom[i] #variable to keep track of domain tested
        
        #variable assignments
        if V=='A': A=d 
        if V=='B': B=d
        if V=='C': C=d 
        if V=='D': D=d 
        if V=='E': E=d 
        if V=='F': F=d 
        if V=='G': G=d 
        if V=='H': H=d 

        if(check_constraints(A,B,C,D,E,F,G,H,Dom)): #if all constraints are met for vals inside domain
            
            path.append(d) #add current variable to path
            
            if(j in range(0, len(ordering) - 1)): #if we still have vars to iterate through, go to next var
                            j+=1
            print(A,B, C, D, E, F, G, H, "success added to path")
            
            if(len(path) != len(ordering)): #if we have not reached a sol yet
                i = 0
                
            else :#(len(path) == len(ordering)): #found a solution
                
                print("***FOUND SOLUTION***")
                print(A,B,C,D,E,F,G,H, "SOLUTION")
                solutions.append('[')
                for X in path:
                    solutions.append(X)
                solutions.append(']')
                print(solutions)
              #  print('path =', path)
                sols.append(path) #add current path to solution array
              #  print("total sols found = ", sols)

                sols_found +=1 #increment number of solutions found
                
                path.pop() #remove last working element to test additional solutions
                
                V = ordering[j] #deassign last element
                if V=='A': A=0
                if V=='B': B=0
                if V=='C': C=0
                if V=='D': D=0
                if V=='E': E=0
                if V=='F': F=0
                if V=='G': G=0
                if V=='H': H=0
                
                if(i in range(0, len(Dom)-1)): #if we have more domains to iterate through on current var
                    i+=1 
                else: 
                    if(j in range(1,len(ordering))): #else backtrack 
                        print('path = ',path)
                        i = path.pop() 
                        j-=1
                        V = ordering[j] #deassign popped element
                        if V=='A': A=0
                        if V=='B': B=0
                        if V=='C': C=0
                        if V=='D': D=0
                        if V=='E': E=0
                        if V=='F': F=0
                        if V=='G': G=0
                        if V=='H': H=0
                            
                        while (i ==4): #backtrack if iterated through all domains                   
                            i = path.pop()
                            V = ordering[j]
                            if V=='A': A=0
                            if V=='B': B=0
                            if V=='C': C=0
                            if V=='D': D=0
                            if V=='E': E=0
                            if V=='F': F=0
                            if V=='G': G=0
                            if V=='H': H=0
                            j-=1
                
        else: #constraints not met, current path fails
            print(A, B, C, D, E, F, G, H, "fail")
            fails_found+=1
            if(i in range(0, len(Dom)-1)): #if we still have domains to iterate through, increment
                i+=1 
            else: 
                if(j in range(1,len(ordering))): #backtrack 
                  #  print('path = ',path)
                    V = ordering[j]
                    if V=='A': A=0
                    if V=='B': B=0
                    if V=='C': C=0
                    if V=='D': D=0
                    if V=='E': E=0
                    if V=='F': F=0
                    if V=='G': G=0
                    if V=='H': H=0
                    
                    i = path.pop() 
                    j-=1
                    
                    while (i ==4):
                        if(len(path) == 0):
                            print('FINISH')
                            print('number of fails = ', fails_found)
                            print("number of sols found = ", sols_found)
                            print("Solutions = ", solutions)
                            break
                        i = path.pop()
                        V = ordering[j]
                        if V=='A': A=0
                        if V=='B': B=0
                        if V=='C': C=0
                        if V=='D': D=0
                        if V=='E': E=0
                        if V=='F': F=0
                        if V=='G': G=0
                        if V=='H': H=0
                        j-=1

                else:
                    print('break')
                    break
            
        
    
print('SOLUTIONS Outside loop')
print(solutions)
print('solutions found = ', sols_found)
print('fails = ', fails_found)
```
